plotting number of lines on one graph

```python
t=np.arange(0.,5.0,0.2)
```

```python
plt.plot(t,t**2, 'b--',label='^2')
plt.plot(t ,t**2.2, 'rs',label='^2.2')
plt.plot(t ,t**2.5, 'g^',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],linewidth=5.0)
plt.xlabel('number')
plt.ylabel('sqaure_of_numbers')
plt.title('sqaure of numbers on graph')
plt.grid()
plt.show()#grid
```

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8 ]
lines=plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth=2.0)
plt.setp(lines[1],color='g',linewidth=2.0)
```

```python
 def f(t):
    return np.exp(-t) *np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(1)#create a figure 
plt.subplot(211)# create 2 rows one column and two figure
plt.grid()
plt.plot(t1,f(t1),'b-')
plt.subplot(212)#accessing figure 2
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```
