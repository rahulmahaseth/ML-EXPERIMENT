```python
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
```

```python
clff=KNeighborsClassifier()
```

```python
def knncomparison(data,n_neighbors):
    '''this function find knn and plot the data'''
    x=data[:,:2]
    y=data[:,2]
    #grid cell
    h=0.02
    cmap_light=ListedColormap(['#FFAAAA','#AAAAFF'])
    cmap_bold=ListedColormap(['#FF0000','#0000FF'])
    #core classifier:knn
    clf=KNeighborsClassifier(n_neighbors)
    clf.fit(x,y)
    x_min,x_max=x[:,0].min()-1,x[:,0].max()+1
    y_min,y_max=x[:,1].min()-1,x[:,1].max()+1
    
    #we create a mesh grid(xmin,y_min)to(x_max,y_max) with 0.02 grid space
    xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    print(xx.shape)
    print(yy.shape)
    #pridict the value(0,1)for each element in the grid
    Z=clf.predict(np.c_[xx.ravel(),yy.ravel()]) 
    # convert the out back to the xx shape (we need it to plot the decission boundry)
    Z = Z.reshape(xx.shape)
    # pcolormesh will plot the (xx,yy) grid with colors according to the values of Z
    # it looks like decision boundry
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)
    # scatter plot of with given points
    plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap_bold)
    #defining scale on both axises
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    # set the title
    plt.title('K value = '+str(n_neighbors))
    
    plt.show()
```

```python
def knn_compa(data,k,):
    x=data[['X1','X2']].values
    y=data['Y'].astype(int).values
    clf=KNeighborsClassifier(n_neighbors=k)
    clf.fit(x,y)
    plot_decision_regions(x,y,clf=clf,legend=2)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('knn with k='+str(k))
```
